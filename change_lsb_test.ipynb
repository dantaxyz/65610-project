```python
import numpy as np
import matplotlib.pyplot as plt
from lib6300.audio import wav_read, wav_write
```

```python
input, fs = wav_read("input.wav")
input = np.array(input)
```

```python
plt.plot(input)
```

```python
MAX_INT = 2**31 - 1
```

```python
input_int = np.floor(input * MAX_INT).astype("int32")
```

```python
plt.plot(input_int)
```

```python
lsbs = [x & 1 for x in input_int]
new_lsbs = [0, 1] * (len(lsbs)//2) + ([0] if len(lsbs)%2 else [])
```

```python
modified = [(x&~1)|b for x, b in zip(input_int, new_lsbs)]
```

```python
wav_write(input_int, fs, "input_int.wav")
```

```python
wav_write(modified, fs, "modified.wav")
```
